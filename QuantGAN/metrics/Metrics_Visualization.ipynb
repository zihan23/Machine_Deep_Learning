```python
data = pd.read_excel(data_dir+'/output.xlsx')

title = 'Original Log Return'

x_data = list(data['Date'])


y1= list(data['log_return'])

trace0 = go.Scatter(
    x = x_data,
    y = y1,
    name = 'original log return',
    line = dict(
        width = 2)
)


df = [trace0]

# Edit the layout
layout = dict(title = title,
              xaxis = dict(title = 'Date',showticklabels=False),
              yaxis = dict(title = 'Log Return'),
              )

fig = dict(data=df, layout=layout)
iplot(fig, filename='styled-line')
```

# ACF Score with f(x) = x

```python
data_mlp = pd.read_csv(data_dir+"/out/MLP_wd0_100x4_12-19-19_10_12/GeneratedDataMLP_wd0_100x4_12-19-19_10_12.csv")
data_lstm = pd.read_csv(data_dir+"/out/LSTM_wd0_100x4_12-19-19_10_25/GeneratedDataLSTM_wd0_100x4_12-19-19_10_25.csv")
data_tcn = pd.read_csv(data_dir+"/out/TCN_wd0_100x4_12-19-19_07_57/GeneratedDataTCN_wd0_100x4_12-19-19_07_57.csv")
```

```python
corr_original  = list()
for i in range(1,5):
    corr_original.append(data['log_return'].autocorr(i))
```

```python
corr_lstm_list = list()
for j in range(16):
    corr_lstm = list()
    for i in range(1,5):
        corr_lstm.append(data_lstm.iloc[:,j+2].autocorr(i))
    corr_lstm_list.append(corr_lstm)
corr_lstm_df = pd.DataFrame(corr_lstm_list)
corr_lstm_avg = corr_lstm_df.mean(axis = 0)
acf_lstm = sum([(x-y)**2 for x,y in zip(corr_original,corr_lstm_avg)])
```
